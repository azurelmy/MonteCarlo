#### 1. Simulating the Underlying Price Paths
We diffuse the underlying price using Geometric Brownian Motion.
\begin{align*}
S(t) = S(0) \exp((\mu - \frac{\sigma^2}{2})t+\sigma \sqrt{t} W_t) \\
dS = \mu Sdt+\sigma SdW_t
\end{align*}
$W_t$ is the Wiener Process, follows $W(0)=0$ and $W_{t+dt}-W_t \sim N(0,dt)$.

(1) Generate a sequence of normally distributed data, i.e., the Brownian Motion increment ($dw_t$)

```python
k = np.random.randn(2520)
plt.hist(k,50)
plt.show()
```

(2) Add volatility = 0.2 to the GBM ($\sigma dW_t$)

```python
k=np.random.randn(2520)*0.2
plt.hist(k,50)
plt.show()
```

(3)(a) Create the GBM (generate one path)

```python
k = np.cumprod(1+np.random.randn(252)*0.2/np.sqrt(252))
plt.plot(k)
plt.show()
```
